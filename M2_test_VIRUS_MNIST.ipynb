```python
# ══════════════════════════════════════════════════════════════════════════════
# QNI-CCP EPSILON SWEEP — VIRUS-MNIST DATASET
# Tests Basic HybridResNet vs ADES QNI-CCP trained model
# ══════════════════════════════════════════════════════════════════════════════
#
# Architecture consistency with ADES training (Doc 26):
#   - Latent space: bridge output (B,16) — sigmoid-scaled
#   - S via classifier-only + z.grad (S≈0 fix for 3-axis Pauli)
#   - Both S and direction unit-normalized
#   - NO torch.no_grad() around quantum calls (backprop circuit)
#   - Centroids from TRAIN loader — matches training geometry
#   - Attack formula: z' = z + ε_q × (S_hat ⊙ dir_hat)
#
# ══════════════════════════════════════════════════════════════════════════════

import torch
import torch.nn as nn
import torch.nn.functional as F
import pennylane as qml
from pennylane.qnn import TorchLayer
import numpy as np
import matplotlib.pyplot as plt
from torchvision import transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
from tqdm import tqdm
import random
import warnings
warnings.filterwarnings("ignore")


# ── 1. CONFIGURATION ─────────────────────────────────────────────────────────

device      = torch.device("cuda" if torch.cuda.is_available() else "cpu")
n_qubits    = 8
q_depth     = 6
q_out_dim   = n_qubits * 3    # 24
num_classes = 10
batch_size  = 32

BASIC_MODEL_PATH = '/home/netsec1/Kathan2/hybrid_resnet_NOGAN.pth'
ADES_MODEL_PATH  = 'qni_ccp_virusmnist_ades.pth'
TEST_DATA_PATH   = '/home/netsec1/Kathan2/virus_MNIST dataset/test'
TRAIN_DATA_PATH  = '/home/netsec1/Kathan2/virus_MNIST dataset/train'

# Paper ε_q = 1/(1+42+6) = 0.0204 for VirusMNIST
# Sweep range covers from paper value up to where model fully degrades
QNICCP_EPSILONS = [0.0, 0.01, 0.0204, 0.05, 0.10, 0.20, 0.30,
                   0.50, 0.80, 1.0, 1.5, 2.0, 2.5, 3.0, 4, 5]

print(f"Device     : {device}")
print(f"Qubits     : {n_qubits}  |  Classes: {num_classes}  |  q_out_dim: {q_out_dim}")
print(f"Paper ε_q  : 0.0204  (1/(1+42+6))")
print(f"ADES range : [{0.0204:.4f}, {0.0204+1.5:.4f}]  (ε_min + λ_max)")


# ── 2. QUANTUM CIRCUIT ────────────────────────────────────────────────────────
# Identical to ADES training — same circuit, same weight_shapes
# diff_method="backprop" required — NO torch.no_grad() around q_layer calls

dev = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(dev, interface="torch", diff_method="backprop")
def quantum_circuit(inputs, weights):
    for i in range(n_qubits):
        qml.RY(inputs[..., i],            wires=i)
        qml.RZ(inputs[..., i + n_qubits], wires=i)

    for l in range(weights.shape[0]):
        if l % 2 == 0:
            for i in range(0, n_qubits - 1, 2):
                qml.CRZ(weights[l, i, 2], wires=[i, i + 1])
        else:
            for i in range(1, n_qubits - 1, 2):
                qml.CRZ(weights[l, i, 2], wires=[i, i + 1])

        for i in range(n_qubits):
            qml.RY(weights[l, i, 0] + inputs[..., i],            wires=i)
            qml.RZ(weights[l, i, 1] + inputs[..., i + n_qubits], wires=i)

    measurements = []
    for i in range(n_qubits):
        measurements.append(qml.expval(qml.PauliZ(i)))
        measurements.append(qml.expval(qml.PauliX(i)))
        measurements.append(qml.expval(qml.PauliY(i)))
    return measurements

weight_shapes = {"weights": (q_depth, n_qubits, 3)}


# ── 3. MODEL ARCHITECTURE ─────────────────────────────────────────────────────
# Identical to ADES training — must match checkpoint exactly

class SEBlock(nn.Module):
    def __init__(self, channels, reduction=4):
        super().__init__()
        self.pool = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Sequential(
            nn.Linear(channels, max(channels // reduction, 4), bias=False),
            nn.ReLU(),
            nn.Linear(max(channels // reduction, 4), channels, bias=False),
            nn.Sigmoid()
        )

    def forward(self, x):
        b, c, _, _ = x.shape
        scale = self.pool(x).view(b, c)
        scale = self.fc(scale).view(b, c, 1, 1)
        return x * scale


class ResBlock(nn.Module):
    def __init__(self, in_ch, out_ch, stride=1, dropout=0.20, drop_path=0.10):
        super().__init__()
        self.conv_block = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, stride=stride, padding=1, bias=False),
            nn.BatchNorm2d(out_ch),
            nn.ReLU(inplace=True),
            nn.Dropout2d(dropout),
            nn.Conv2d(out_ch, out_ch, 3, stride=1, padding=1, bias=False),
            nn.BatchNorm2d(out_ch)
        )
        self.se             = SEBlock(out_ch)
        self.drop_path_rate = drop_path
        self.skip = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 1, stride=stride, bias=False),
            nn.BatchNorm2d(out_ch)
        ) if (stride != 1 or in_ch != out_ch) else nn.Identity()
        self.relu = nn.ReLU(inplace=True)

    def forward(self, x):
        out = self.conv_block(x)
        out = self.se(out)
        if self.training and self.drop_path_rate > 0:
            keep_prob     = 1 - self.drop_path_rate
            random_tensor = (
                torch.rand(x.shape[0], 1, 1, 1, device=x.device) < keep_prob
            ).float()
            out = out * random_tensor / keep_prob
        return self.relu(out + self.skip(x))


class QuantumBridge(nn.Module):
    def __init__(self, in_features, n_qubits):
        super().__init__()
        self.project = nn.Sequential(
            nn.Linear(in_features, 32),
            nn.LayerNorm(32),
            nn.GELU(),
            nn.Dropout(0.35),
            nn.Linear(32, n_qubits * 2)
        )
        self.angle_scale = nn.Parameter(torch.ones(n_qubits * 2) * torch.pi)
        self.angle_bias  = nn.Parameter(torch.zeros(n_qubits * 2))

    def forward(self, x):
        x = self.project(x)
        return self.angle_scale * torch.sigmoid(x) + self.angle_bias


class HybridResNet(nn.Module):
    """
    Identical architecture to ADES training.
    Both Basic and ADES models use this same class — only weights differ.

    QNI-CCP sweep-specific methods:
      get_bridge_features(): extracts (B,16) bridge output for centroid/attack
      sensitivity_from_classifier(): ∂Loss/∂z via classifier only
        - S≈0 fix: 3-axis Pauli measurement collapses gradients through q_layer
        - Uses z.grad directly (true gradient at bridge output)
        - Clears model grads before/after to avoid contaminating sweep
      forward_from_bridge(): evaluates logits from perturbed bridge features
    """
    def __init__(self, n_qubits, q_out_dim, num_classes, dropout=0.35):
        super().__init__()
        self.stem = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=3, stride=1, padding=1, bias=False),
            nn.BatchNorm2d(16),
            nn.ReLU(inplace=True)
        )
        self.stage1 = nn.Sequential(
            ResBlock(16, 16, drop_path=0.10),
            ResBlock(16, 16, drop_path=0.10)
        )
        self.stage2 = nn.Sequential(
            ResBlock(16, 32, stride=2, drop_path=0.15),
            ResBlock(32, 32,           drop_path=0.15)
        )
        self.stage3 = nn.Sequential(
            ResBlock(32, 64, stride=2, drop_path=0.20),
            ResBlock(64, 64,           drop_path=0.20)
        )
        self.gap        = nn.AdaptiveAvgPool2d(1)
        self.bridge     = QuantumBridge(in_features=64, n_qubits=n_qubits)
        self.q_layer    = TorchLayer(quantum_circuit, weight_shapes)
        self.classifier = nn.Sequential(
            nn.Linear(q_out_dim, q_out_dim * 2),
            nn.ReLU(inplace=True),
            nn.Dropout(dropout),
            nn.Linear(q_out_dim * 2, num_classes)
        )
        self._init_weights()

    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
            elif isinstance(m, (nn.BatchNorm2d, nn.LayerNorm)):
                nn.init.ones_(m.weight)
                nn.init.zeros_(m.bias)
            elif isinstance(m, nn.Linear):
                nn.init.kaiming_normal_(m.weight)
                if m.bias is not None:
                    nn.init.zeros_(m.bias)

    def _get_backbone_features(self, x):
        x = self.stem(x)
        x = self.stage1(x)
        x = self.stage2(x)
        x = self.stage3(x)
        x = self.gap(x)
        return x.view(x.size(0), -1)   # (B, 64)
    def get_cnn_features(self, x):
        """CNN only → (B, 64) — QNI-CCP latent space."""
        with torch.no_grad():
            return self._get_backbone_features(x)
    def get_bridge_features(self, x):
        """
        CNN + bridge → (B, 16) bridge output.
        torch.no_grad() is safe here — no quantum circuit involved.
        Used for centroid computation and clean z extraction in attack.
        """
        with torch.no_grad():
            backbone = self._get_backbone_features(x)   # (B, 64)
            return self.bridge(backbone)                 # (B, 16)

    def sensitivity_from_classifier(self, z_bridge, labels):
        """
        S = ∂Loss/∂z_bridge in bridge output space (B, 16).
    
        Bug 2 was: using W1[:, :16] — but classifier[0] expects 24-dim input
        (q_layer output), not 16. The first 16 columns correspond to the first
        24 q_layer features, which aren't the bridge features themselves.
    
        Fix: use the full W1 (48×24) but treat bridge output as a proxy for
        q_layer output by padding/mapping it. Simpler and more correct: compute
        gradient directly on z_bridge using bridge→linear_proxy chain where
        the linear proxy maps 16→24 (identity pad or learned projection).
    
        The most robust fix is to use FGSM directly on z_bridge:
        just take sign(∂CE/∂z_bridge) through bridge→classifier with
        q_layer replaced by identity projection.
        """
        for p in self.parameters():
            if p.grad is not None:
                p.grad.detach_()
                p.grad.zero_()
    
        z_leaf = z_bridge.clone().detach().requires_grad_(True)  # (B, 16)
    
        # Proxy: map bridge output (16) → q_layer output size (24) via linear pad
        # This avoids quantum circuit and gives real gradient signal
        # Use learned q_layer weight approximation: first linear of classifier
        # expects 24 inputs — we pad bridge output to match
        pad = torch.zeros(z_leaf.size(0), 8, device=z_leaf.device)
        z_padded = torch.cat([z_leaf, pad], dim=1)  # (B, 24)
    
        W1 = self.classifier[0].weight.detach()   # (48, 24) — FULL matrix, not sliced
        b1 = self.classifier[0].bias.detach()
        W2 = self.classifier[-1].weight.detach()  # (10, 48)
        b2 = self.classifier[-1].bias.detach()
    
        h      = F.relu(z_padded @ W1.T + b1)
        logits = h @ W2.T + b2
        loss   = F.cross_entropy(logits, labels)
        loss.backward()
    
        S = z_leaf.grad.data.clone() if z_leaf.grad is not None else \
            torch.ones_like(z_bridge)
    
        for p in self.parameters():
            if p.grad is not None:
                p.grad.detach_()
                p.grad.zero_()
    
        return S / (S.norm(dim=1, keepdim=True) + 1e-8)  # (B, 16) unit-norm
    def forward(self, x):
        x = self._get_backbone_features(x)   # (B, 64)
        z = self.bridge(x)                    # (B, 16)
        q = self.q_layer(z)                   # (B, 24)
        return self.classifier(q)


# ── 4. DATASET ────────────────────────────────────────────────────────────────
# 32×32 — identical to ADES training transforms

eval_transform = transforms.Compose([
    transforms.Grayscale(1),
    transforms.Resize((32, 32)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

test_dataset  = ImageFolder(TEST_DATA_PATH,  transform=eval_transform)
train_dataset = ImageFolder(TRAIN_DATA_PATH, transform=eval_transform)

test_loader  = DataLoader(test_dataset,  batch_size=batch_size,
                          shuffle=False, num_workers=4, pin_memory=True)
train_loader = DataLoader(train_dataset, batch_size=batch_size,
                          shuffle=False, num_workers=4, pin_memory=True)

print(f"\nTest  samples : {len(test_dataset)}")
print(f"Train samples : {len(train_dataset)}  ← used for centroids only")
print(f"Classes found : {len(test_dataset.classes)}")


# ── 5. MODEL LOADING ──────────────────────────────────────────────────────────

def load_model(path, label):
    """
    Loads HybridResNet from checkpoint.
    Both Basic and ADES models share the same architecture.
    Only weights differ — Basic=clean training, ADES=robustness training.
    """
    model = HybridResNet(
        n_qubits    = n_qubits,
        q_out_dim   = q_out_dim,
        num_classes = num_classes,
        dropout     = 0.35
    ).to(device)

    ckpt  = torch.load(path, map_location=device)
    state = ckpt['model_state_dict'] if isinstance(ckpt, dict) and \
            'model_state_dict' in ckpt else ckpt
    model.load_state_dict(state)
    model.eval()

    val_acc = ckpt.get('val_acc', '?') if isinstance(ckpt, dict) else '?'
    print(f"✅ Loaded [{label}] from: {path}")
    print(f"   Saved val_acc: {val_acc}")
    return model

basic_model = load_model(BASIC_MODEL_PATH, "Basic HybridResNet — clean baseline")
ades_model  = load_model(ADES_MODEL_PATH,  "ADES QNI-CCP HybridResNet")


# ── 6. CENTROID COMPUTATION ───────────────────────────────────────────────────

def compute_class_centroids_bridge(model, loader, device, num_classes):
    """
    μ_c = mean BRIDGE output per class — shape (10, 16).
    Bridge space is where perturbation actually happens (after the fix).
    No periodicity issue because we perturb z_bridge directly and feed
    it straight to q_layer — the bridge sigmoid is NOT re-applied.
    """
    bridge_dim = 16
    model.eval()
    sum_features = torch.zeros(num_classes, bridge_dim, device=device)
    count        = torch.zeros(num_classes, device=device)

    with torch.no_grad():
        for x, y in tqdm(loader, desc="Computing bridge centroids", leave=False):
            x, y     = x.to(device), y.to(device)
            features = model.get_bridge_features(x)  # (B, 16)
            for c in range(num_classes):
                mask = (y == c)
                if mask.any():
                    sum_features[c] += features[mask].sum(dim=0)
                    count[c]        += mask.sum()

    count[count == 0] = 1.0
    return sum_features / count.unsqueeze(1)  # (10, 16)

# ── 7. QNI-CCP ATTACK ─────────────────────────────────────────────────────────

def qni_ccp_attack(model, images, labels, epsilon_q, centroids):
    """
    Correct QNI-CCP: perturbation happens in IMAGE space (B,1,32,32).
    
    The quantum sensitivity S tells us WHICH directions in image space
    are most sensitive to misclassification. We perturb pixels, not
    the latent vector — that's what makes this a real adversarial attack.
    """
    if epsilon_q == 0.0:
        with torch.no_grad():
            return model(images).detach()

    model.eval()

    # Step 1: get bridge centroids for direction guidance
    with torch.no_grad():
        z_clean = model.get_bridge_features(images)  # (B, 16)

    # Step 2: wrong-class centroid direction in bridge space
    target_classes = []
    for i in range(labels.size(0)):
        available = [c for c in range(centroids.size(0)) if c != labels[i].item()]
        target_classes.append(
            random.choice(available) if available
            else (labels[i].item() + 1) % centroids.size(0)
        )
    target_tensor = torch.tensor(target_classes, device=device)
    mu_wrong      = centroids[target_tensor]  # (B, 16)

    direction = mu_wrong - z_clean
    dir_hat   = direction / (direction.norm(dim=1, keepdim=True) + 1e-8)  # (B, 16)

    # Step 3: compute gradient of loss w.r.t. INPUT IMAGE
    # This is where QNI-CCP differs from plain FGSM — the loss
    # combines cross-entropy AND centroid pull in bridge space
    x_adv = images.clone().detach().requires_grad_(True)

    # Full forward pass with gradient tracking through everything
    backbone   = model._get_backbone_features(x_adv)  # (B, 64)
    z_bridge   = model.bridge(backbone)                # (B, 16)
    q_out      = model.q_layer(z_bridge)               # (B, 24)
    logits     = model.classifier(q_out)               # (B, 10)

    # Combined loss: CE pushes toward wrong class + centroid pull
    loss_ce       = F.cross_entropy(logits, labels)
    loss_centroid = -F.mse_loss(z_bridge, mu_wrong)    # negative = pull toward wrong centroid
    
    # Weight the centroid pull by sensitivity direction
    sensitivity_loss = -(z_bridge * dir_hat).sum(dim=1).mean()
    
    total_loss = loss_ce + 0.5 * sensitivity_loss
    total_loss.backward()

    # Step 4: FGSM step in image space — sign of gradient
    with torch.no_grad():
        grad_sign = x_adv.grad.data.sign()             # (B, 1, 32, 32)
        x_perturbed = images + epsilon_q * grad_sign   # perturb pixels
        # Clamp to valid normalized range [-1, 1] (your normalization uses mean=0.5, std=0.5)
        x_perturbed = x_perturbed.clamp(-1.0, 1.0)

    # Step 5: forward on perturbed image — this is the real attacked output
    with torch.no_grad():
        logits_adv = model(x_perturbed)

    return logits_adv.detach()





# ── 8. EVALUATION FUNCTION ────────────────────────────────────────────────────

def evaluate_epsilon_fixed(model, loader, centroids, epsilon_q, label):
    model.eval()
    all_preds  = []
    all_labels = []

    for images, labels in tqdm(loader,
                               desc=f"  ε={epsilon_q:.4f} | {label}",
                               leave=False):
        images, labels = images.to(device), labels.to(device)
        logits = qni_ccp_attack(model, images, labels,
                                       epsilon_q, centroids)
        preds  = logits.argmax(dim=1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

    acc  = accuracy_score(all_labels, all_preds)
    prec = precision_score(all_labels, all_preds, average='macro', zero_division=0)
    rec  = recall_score(all_labels,   all_preds, average='macro', zero_division=0)
    f1   = f1_score(all_labels,       all_preds, average='macro', zero_division=0)
    return acc, prec, rec, f1

# ── 9. COMPUTE CENTROIDS ──────────────────────────────────────────────────────

print("\nComputing bridge-space centroids for Basic model...")
centroids_basic = compute_class_centroids_bridge(basic_model, train_loader,
                                                  device, num_classes)
print("Computing bridge-space centroids for ADES model...")
centroids_ades  = compute_class_centroids_bridge(ades_model,  train_loader,
                                                  device, num_classes)
print(f"✅ Centroids ready. Shape: {centroids_basic.shape}")  # (10, 16)


# ── 10. EPSILON SWEEP ─────────────────────────────────────────────────────────

results = {
    "Basic" : {"acc": [], "prec": [], "rec": [], "f1": []},
    "ADES"  : {"acc": [], "prec": [], "rec": [], "f1": []},
}

print("═"*65)
print("  QNI-CCP EPSILON SWEEP — VIRUS-MNIST | Basic vs ADES")
print("═"*65)

for eps in QNICCP_EPSILONS:
    print(f"\n[ε_q = {eps:.4f}]")

    acc, prec, rec, f1 = evaluate_epsilon(
        basic_model, test_loader, centroids_basic, eps, "Basic"
    )
    results["Basic"]["acc"].append(acc)
    results["Basic"]["prec"].append(prec)
    results["Basic"]["rec"].append(rec)
    results["Basic"]["f1"].append(f1)
    print(f"  Basic | Acc:{acc:.4f} Prec:{prec:.4f} Rec:{rec:.4f} F1:{f1:.4f}")

    acc, prec, rec, f1 = evaluate_epsilon(
        ades_model, test_loader, centroids_ades, eps, "ADES"
    )
    results["ADES"]["acc"].append(acc)
    results["ADES"]["prec"].append(prec)
    results["ADES"]["rec"].append(rec)
    results["ADES"]["f1"].append(f1)
    print(f"  ADES  | Acc:{acc:.4f} Prec:{prec:.4f} Rec:{rec:.4f} F1:{f1:.4f}")


# ── 11. SUMMARY TABLE ─────────────────────────────────────────────────────────

print(f"\n{'═'*80}")
print(f"  SUMMARY — QNI-CCP ε_q SWEEP  |  Virus-MNIST  |  Basic vs ADES")
print(f"{'═'*80}")
print(f"  {'ε_q':>7}   {'Basic Acc':>9} {'ADES Acc':>9}   "
      f"{'Basic F1':>9} {'ADES F1':>9}   {'Δ Acc':>8} {'Δ F1':>8}")
print(f"  {'─'*75}")

for i, eps in enumerate(QNICCP_EPSILONS):
    b_acc = results["Basic"]["acc"][i]
    a_acc = results["ADES"]["acc"][i]
    b_f1  = results["Basic"]["f1"][i]
    a_f1  = results["ADES"]["f1"][i]
    d_acc = a_acc - b_acc   # positive = ADES more robust
    d_f1  = a_f1  - b_f1
    s_acc = "↑" if d_acc >= 0 else "↓"
    s_f1  = "↑" if d_f1  >= 0 else "↓"
    marker = " ◄ paper ε_q" if abs(eps - 0.0204) < 1e-4 else ""
    print(f"  {eps:>7.4f}   {b_acc:>9.4f} {a_acc:>9.4f}   "
          f"{b_f1:>9.4f} {a_f1:>9.4f}   "
          f"{d_acc:>+7.4f}{s_acc} {d_f1:>+7.4f}{s_f1}{marker}")


# ── 12. PLOTS ─────────────────────────────────────────────────────────────────

metrics      = ["acc",      "prec",      "rec",    "f1"]
metric_names = ["Accuracy", "Precision", "Recall", "F1-Score (Macro)"]
colors       = {"Basic": "#E74C3C", "ADES": "#2980B9"}

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle(
    "QNI-CCP ε_q Sweep: Basic HybridResNet vs ADES QNI-CCP — Virus-MNIST\n"
    "Bridge-space perturbation strength vs classification performance",
    fontsize=13, fontweight='bold'
)
axes = axes.flatten()

for idx, (metric, mname) in enumerate(zip(metrics, metric_names)):
    ax         = axes[idx]
    basic_vals = results["Basic"][metric]
    ades_vals  = results["ADES"][metric]

    ax.plot(QNICCP_EPSILONS, basic_vals, marker='o', color=colors["Basic"],
            label='Basic HybridResNet', linewidth=2, markersize=6)
    ax.plot(QNICCP_EPSILONS, ades_vals,  marker='s', color=colors["ADES"],
            label='ADES QNI-CCP',       linewidth=2, markersize=6, linestyle='--')

    # Paper ε_q for VirusMNIST
    ax.axvline(x=0.0204, color='gray', linestyle=':', linewidth=1.5,
               label='Paper ε_q=0.0204')

    # Robustness gap — green where ADES outperforms Basic
    ax.fill_between(QNICCP_EPSILONS, basic_vals, ades_vals,
                    alpha=0.08, color='green', label='Robustness gap')

    ax.set_title(mname, fontsize=11, fontweight='bold')
    ax.set_xlabel("QNI-CCP Epsilon (ε_q)", fontsize=9)
    ax.set_ylabel(mname, fontsize=9)
    ax.set_ylim(-0.05, 1.05)
    ax.legend(fontsize=8, loc='lower left')
    ax.grid(True, alpha=0.3)
    ax.set_xticks(QNICCP_EPSILONS)
    ax.tick_params(axis='x', rotation=45, labelsize=7)

plt.tight_layout()
plt.savefig("qniccp_sweep_virusmnist_ades.png", dpi=150, bbox_inches='tight')
plt.show()
print("✅ Plot saved: qniccp_sweep_virusmnist_ades.png")


# ── 13. SAVE RESULTS ──────────────────────────────────────────────────────────

with open("qniccp_sweep_virusmnist_ades_results.txt", "w") as f:
    f.write("QNI-CCP EPSILON SWEEP RESULTS — VIRUS-MNIST\n")
    f.write("="*80 + "\n\n")
    f.write(f"{'eps_q':>7}  {'Model':>12}  "
            f"{'Accuracy':>10} {'Precision':>10} {'Recall':>10} {'F1':>10}\n")
    f.write("-"*65 + "\n")
    for i, eps in enumerate(QNICCP_EPSILONS):
        for key, lbl in [("Basic", "Basic"), ("ADES", "ADES QNI-CCP")]:
            f.write(f"{eps:>7.4f}  {lbl:>12}  "
                    f"{results[key]['acc'][i]:>10.4f} "
                    f"{results[key]['prec'][i]:>10.4f} "
                    f"{results[key]['rec'][i]:>10.4f} "
                    f"{results[key]['f1'][i]:>10.4f}\n")

print("✅ Results saved: qniccp_sweep_virusmnist_ades_results.txt")
```
