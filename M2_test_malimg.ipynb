```python
# ============================================================
# ADES QNI-CCP EPSILON SWEEP — MALIMG DATASET
# Evaluates Basic QNN vs ADES-trained model under QNI-CCP attack.
# Exact same sweep structure as your previous Malimg sweep code.
# Architecture matches qni_ccp_maleimg2.pth checkpoint exactly:
#   - n_qubits=6, num_classes=25, image 224×224 (your ADES training)
#   - FeatureReduce: single fc layer (128→6)
#   - quantum_circuit: 3 layers, torch.stack forward (no backprop)
#   - Sensitivity: classifier-only path (same reason as before)
# ============================================================

import torch
import torch.nn as nn
import torch.nn.functional as F
import pennylane as qml
from pennylane.qnn import TorchLayer
import numpy as np
import matplotlib.pyplot as plt
from torchvision import transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
from tqdm.notebook import tqdm
import random
import warnings
warnings.filterwarnings("ignore")

# ── 1. CONFIGURATION ────────────────────────────────────────────

device      = torch.device("cuda" if torch.cuda.is_available() else "cpu")
n_qubits    = 6       # Matches your ADES training config exactly
num_classes = 25      # Malimg: 25 families
batch_size  = 16

# Basic model: your original Malimg baseline
# ADES model: saved from your ADES training run
BASIC_MODEL_PATH = "/home/netsec1/Kathan/best_model_final_ch4.pth"
ADES_MODEL_PATH  = "qni_ccp_ades_malimg.pth"
TEST_DATA_PATH   = "/home/netsec1/Kathan/dataset_folder/malimg_dataset/test"
TRAIN_DATA_PATH   = "/home/netsec1/Kathan/dataset_folder/malimg_dataset/train"



# Sweep range — same as your previous Malimg sweeps for direct comparison
# Paper ε_q ≈ 0.027 for Malimg (1/(1+30+6))
# ADES epsilon_min=0.0164, lambda_max=2.5 → range [0.0164, 2.5164]
QNICCP_EPSILONS = [0.0, 0.01, 0.027,0.04, 0.05, 0.07, 0.10, 0.20, 0.25, 0.30,
                   0.50, 0.80, 1.0, 1.5, 2.0]

print(f"Device     : {device}")
print(f"Qubits     : {n_qubits}  |  Classes: {num_classes}")
print(f"Paper ε_q  : 0.027  (Malimg — 1/(1+30+6))")
print(f"ADES range : [{0.0164:.4f}, {0.0164+2.5:.4f}]")


# ── 2. QUANTUM CIRCUITS ─────────────────────────────────────────
# Basic model: 3 variational layers (matches best_model_final_ch4.pth)
# ADES model:  3 variational layers (matches your ADES training weight_shapes)
# Both use interface="torch" WITHOUT backprop — same as your ADES training.
# Sensitivity computed through classifier-only path for both models.

dev_basic = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(dev_basic, interface="torch")
def quantum_circuit_basic(inputs, weights):
    # Angle embedding — per-sample (no batch mode, matches torch.stack forward)
    for i in range(n_qubits):
        qml.RY(inputs[i], wires=i)          # inputs[i] = scalar per qubit
    for l in range(weights.shape[0]):        # 3 layers for basic model
        for i in range(n_qubits):
            qml.RY(weights[l][i], wires=i)
        for i in range(n_qubits - 1):
            qml.CNOT(wires=[i, i+1])
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

weight_shapes_basic = {"weights": (3, n_qubits)}   # 3 layers × 6 qubits


dev_ades = qml.device("default.qubit", wires=n_qubits)

@qml.qnode(dev_ades, interface="torch")
def quantum_circuit_ades(inputs, weights):
    # Identical circuit to basic — ADES training used same 3-layer structure
    # The difference is in the WEIGHTS (trained with ADES perturbation)
    # not in the circuit architecture itself
    for i in range(n_qubits):
        qml.RY(inputs[i], wires=i)
    for l in range(weights.shape[0]):        # 3 layers — matches weight_shapes
        for i in range(n_qubits):
            qml.RY(weights[l][i], wires=i)
        for i in range(n_qubits - 1):
            qml.CNOT(wires=[i, i+1])
    return [qml.expval(qml.PauliZ(i)) for i in range(n_qubits)]

weight_shapes_ades = {"weights": (3, n_qubits)}    # 3 layers × 6 qubits


# ── 3. MODEL ARCHITECTURE ───────────────────────────────────────
# Must match your ADES training code exactly.
# FeatureReduce: single fc layer (128→6) — Malimg architecture.
# HybridQNN: torch.stack forward — same as your ADES training HybridQNN.

class FeatureReduce(nn.Module):
    """
    CNN: 224×224 grayscale → 6-dim latent vector.
    Single fc layer (128→6) — Malimg architecture.
    Dropout=0.4 in CNN encoder — matches your ADES training.
    Note: your ADES training resized to 224×224 (not 128×128),
    so we use 224×224 here to match the checkpoint.
    """
    def __init__(self, final_dim, dropout=0.4):
        super().__init__()
        self.conv = nn.Sequential(
            nn.Conv2d(1, 8, 3, stride=2, padding=1),
            nn.BatchNorm2d(8), nn.ReLU(), nn.Dropout(dropout),

            nn.Conv2d(8, 16, 3, stride=2, padding=1),
            nn.BatchNorm2d(16), nn.ReLU(), nn.Dropout(dropout),

            nn.Conv2d(16, 32, 3, stride=2, padding=1),
            nn.BatchNorm2d(32), nn.ReLU(), nn.Dropout(dropout),

            nn.Conv2d(32, 64, 3, stride=2, padding=1),
            nn.BatchNorm2d(64), nn.ReLU(), nn.Dropout(dropout),

            nn.Conv2d(64, 128, 3, stride=2, padding=1),
            nn.BatchNorm2d(128), nn.ReLU(),

            nn.AdaptiveAvgPool2d((1, 1))    # → [batch, 128, 1, 1]
        )
        self.fc = nn.Linear(128, final_dim)   # Single FC: 128→6

    def forward(self, x):
        x = self.conv(x)              # [batch, 128, 1, 1]
        x = x.view(x.size(0), -1)    # [batch, 128]
        return self.fc(x)             # [batch, 6] raw pre-tanh features


class HybridQNN_Basic(nn.Module):
    """
    Basic Malimg model — 3-layer quantum circuit, 6 qubits, 25 classes.
    torch.stack forward — matches best_model_final_ch4.pth exactly.
    Classifier dropout=0.3 — matches basic model training.
    get_features(): returns tanh(CNN(x)) for centroid/attack computation.
    sensitivity_from_classifier(): computes S through classifier only.
    Reason: interface="torch" without backprop blocks quantum gradients.
    forward_from_perturbed(): evaluates logits from attacked features.
    """
    def __init__(self, n_qubits, num_classes):
        super().__init__()
        self.feature_extractor = FeatureReduce(final_dim=n_qubits)
        self.q_layer           = TorchLayer(quantum_circuit_basic,
                                            weight_shapes_basic)
        self.classifier        = nn.Sequential(
            nn.Linear(n_qubits, 64), nn.ReLU(), nn.Dropout(0.3),
            nn.Linear(64, 32),       nn.ReLU(), nn.Dropout(0.3),
            nn.Linear(32, 16),       nn.ReLU(),
            nn.Linear(16, num_classes)
        )

    def forward(self, x):
        x     = torch.tanh(self.feature_extractor(x))          # [batch, 6]
        q_out = torch.stack([self.q_layer(f) for f in x])      # [batch, 6]
        return self.classifier(q_out)                          # [batch, 25]

    def get_features(self, x):
        """tanh-activated CNN features — latent space z for attack."""
        with torch.no_grad():
            return torch.tanh(self.feature_extractor(x))       # [batch, 6]

    def sensitivity_from_classifier(self, features, labels):
        """
        S = ∂Loss/∂features through classifier only (skip quantum).
        Why: interface="torch" without diff_method blocks gradient
        propagation through quantum gates back to input features.
        Classifier-only path is fully differentiable and gives
        a valid approximation of feature sensitivity.
        """
        logits = self.classifier(features)          # [batch, 25]
        loss   = F.cross_entropy(logits, labels)    # Scalar
        loss.backward()
        return features.grad.data.clone()           # [batch, 6]

    def forward_from_perturbed(self, z_prime):
        """Evaluate logits from perturbed features — no CNN, no grad."""
        with torch.no_grad():
            q_out = torch.stack([self.q_layer(f) for f in z_prime])
            return self.classifier(q_out)           # [batch, 25]


class HybridQNN_ADES(nn.Module):
    """
    ADES-trained Malimg model — 3-layer quantum circuit, 6 qubits, 25 classes.
    Identical architecture to Basic — only weights differ.
    Classifier dropout=0.3 — matches your ADES training HybridQNN.
    Same helper methods as Basic for consistent attack evaluation.
    The ADES model's classifier was trained with adaptive perturbation,
    so its sensitivity landscape S should be smoother/smaller → slower
    degradation under attack → positive robustness gap vs Basic.
    """
    def __init__(self, n_qubits, num_classes):
        super().__init__()
        self.feature_extractor = FeatureReduce(final_dim=n_qubits)
        self.q_layer           = TorchLayer(quantum_circuit_ades,
                                            weight_shapes_ades)
        self.classifier        = nn.Sequential(
            nn.Linear(n_qubits, 64), nn.ReLU(), nn.Dropout(0.3),
            nn.Linear(64, 32),       nn.ReLU(), nn.Dropout(0.3),
            nn.Linear(32, 16),       nn.ReLU(),
            nn.Linear(16, num_classes)
        )

    def forward(self, x):
        x     = torch.tanh(self.feature_extractor(x))
        q_out = torch.stack([self.q_layer(f) for f in x])
        return self.classifier(q_out)

    def get_features(self, x):
        """tanh-activated CNN features — latent space z."""
        with torch.no_grad():
            return torch.tanh(self.feature_extractor(x))

    def sensitivity_from_classifier(self, features, labels):
        """
        Classifier-only sensitivity — same reason as Basic.
        ADES-trained classifier has smoother gradient landscape
        due to adaptive per-sample perturbation during training.
        This means S is smaller → weaker attack effect on ADES model
        → demonstrates ADES robustness advantage over Basic.
        """
        logits = self.classifier(features)
        loss   = F.cross_entropy(logits, labels)
        loss.backward()
        return features.grad.data.clone()

    def forward_from_perturbed(self, z_prime):
        """Forward from perturbed features — evaluation only."""
        with torch.no_grad():
            q_out = torch.stack([self.q_layer(f) for f in z_prime])
            return self.classifier(q_out)


# ── 4. DATASET ──────────────────────────────────────────────────
# 224×224 — matches your ADES training transform exactly
# (your ADES training used 224×224, not the original 128×128)

eval_transform = transforms.Compose([
    transforms.Grayscale(1),
    transforms.Resize((128, 128)),          # Must match ADES training
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])

test_dataset = ImageFolder(TEST_DATA_PATH, transform=eval_transform)
test_loader  = DataLoader(test_dataset, batch_size=batch_size,
                          shuffle=False, num_workers=4)

train_dataset = ImageFolder(TRAIN_DATA_PATH, transform=eval_transform)
train_loader  = DataLoader(train_dataset, batch_size=batch_size,
                          shuffle=False, num_workers=4)
print(f"\nTest samples : {len(test_dataset)}")
print(f"Classes found: {len(test_dataset.classes)}")


# ── 5. MODEL LOADING ─────────────────────────────────────────────

def load_basic_model(path):
    """Load Malimg basic model — 3-layer QNN, 6 qubits, 25 classes."""
    model = HybridQNN_Basic(n_qubits=n_qubits,
                            num_classes=num_classes).to(device)
    ckpt  = torch.load(path, map_location=device)
    state = ckpt['model_state_dict'] if isinstance(ckpt, dict) and \
            'model_state_dict' in ckpt else ckpt
    model.load_state_dict(state)
    model.eval()
    print(f"✅ Loaded [Basic QNN] from: {path}")
    return model

def load_ades_model(path):
    """Load ADES-trained Malimg model — same architecture, ADES weights."""
    model = HybridQNN_ADES(n_qubits=n_qubits,
                           num_classes=num_classes).to(device)
    ckpt  = torch.load(path, map_location=device)
    state = ckpt['model_state_dict'] if isinstance(ckpt, dict) and \
            'model_state_dict' in ckpt else ckpt
    model.load_state_dict(state)
    model.eval()
    print(f"✅ Loaded [ADES QNI-CCP] from: {path}")
    return model

basic_model = load_basic_model(BASIC_MODEL_PATH)
ades_model  = load_ades_model(ADES_MODEL_PATH)


# ── 6. CENTROID COMPUTATION ──────────────────────────────────────

def compute_class_centroids(model, loader, device, num_classes, n_qubits):
    """
    Compute μ_c = mean tanh(CNN(x)) for all x in class c.
    Computed separately per model — different weight → different latent space.
    Shape: [25, 6] for Malimg.
    """
    model.eval()
    sums   = torch.zeros(num_classes, n_qubits, device=device)
    counts = torch.zeros(num_classes, device=device)

    with torch.no_grad():
        for x, y in loader:
            x, y     = x.to(device), y.to(device)
            features = model.get_features(x)    # [batch, 6]
            for c in range(num_classes):
                mask = (y == c)
                if mask.any():
                    sums[c]   += features[mask].sum(0)
                    counts[c] += mask.sum()

    counts[counts == 0] = 1
    return sums / counts.unsqueeze(1)           # [25, 6]


# ── 7. QNI-CCP ATTACK ───────────────────────────────────────────

def qni_ccp_attack(model, images, labels, epsilon_q, centroids):
    """
    Fixed-epsilon QNI-CCP attack for sweep evaluation.
    Formula: z' = z + ε_q · [S ⊙ (μ_c' − z)]

    Uses classifier-only sensitivity (same reason as training):
    interface="torch" blocks quantum gradient propagation.
    This is consistent across Basic and ADES models — fair comparison.

    At ε_q=0.0: clean baseline — model() called directly.
    At higher ε_q: features pushed toward wrong centroid.
    ADES model expected to degrade slower → positive robustness gap.
    """
    if epsilon_q == 0.0:
        with torch.no_grad():
            return model(images)       # Clean forward pass

    model.eval()

    # Step 1: Clean latent features
    original_features = model.get_features(images)     # [batch, 6]

    # Step 2: Sensitivity S — classifier-only path
    features_for_grad = original_features.clone().detach().requires_grad_(True)
    S = model.sensitivity_from_classifier(features_for_grad, labels)
    # S: [batch, 6]

    # Step 3: Random wrong-class target centroid
    target_classes = []
    for i in range(labels.size(0)):
        available = [c for c in range(centroids.size(0))
                     if c != labels[i].item()]
        target_classes.append(
            random.choice(available) if available
            else (labels[i].item() + 1) % centroids.size(0)
        )
    target_tensor = torch.tensor(target_classes, device=device)
    mu_c_prime    = centroids[target_tensor]            # [batch, 6]

    # Step 4: Apply perturbation
    direction          = mu_c_prime - original_features  # [batch, 6]
    weighted_direction = S * direction                   # [batch, 6]
    z_prime            = original_features + \
                         epsilon_q * weighted_direction  # [batch, 6]

    # Step 5: Evaluate from perturbed features
    logits = model.forward_from_perturbed(z_prime)      # [batch, 25]
    return logits


# ── 8. EVALUATION FUNCTION ──────────────────────────────────────

def evaluate_epsilon(model, loader, centroids, epsilon_q, label):
    """
    Evaluate model under QNI-CCP attack at given epsilon_q.
    Macro averaging — treats all 25 Malimg families equally.
    Malimg is imbalanced so macro avg is the fair metric.
    """
    model.eval()
    all_preds  = []
    all_labels = []

    for images, labels in tqdm(loader,
                               desc=f"  ε={epsilon_q:.3f} | {label}",
                               leave=False):
        images, labels = images.to(device), labels.to(device)
        logits = qni_ccp_attack(model, images, labels, epsilon_q, centroids)
        preds  = logits.argmax(dim=1)
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

    acc  = accuracy_score(all_labels, all_preds)
    prec = precision_score(all_labels, all_preds,
                           average='macro', zero_division=0)
    rec  = recall_score(all_labels, all_preds,
                        average='macro', zero_division=0)
    f1   = f1_score(all_labels, all_preds,
                    average='macro', zero_division=0)
    return acc, prec, rec, f1


# ── 9. COMPUTE CENTROIDS ────────────────────────────────────────

print("\nComputing centroids for Basic model...")
centroids_basic = compute_class_centroids(basic_model, train_loader,
                                          device, num_classes, n_qubits)
print("Computing centroids for ADES model...")
centroids_ades  = compute_class_centroids(ades_model,  train_loader,
                                          device, num_classes, n_qubits)
print("✅ Centroids ready.\n")


# ── 10. EPSILON SWEEP ───────────────────────────────────────────

results = {
    "Basic" : {"acc": [], "prec": [], "rec": [], "f1": []},
    "ADES"  : {"acc": [], "prec": [], "rec": [], "f1": []},
}

print("═"*65)
print("  QNI-CCP EPSILON SWEEP — MALIMG | Basic vs ADES")
print("═"*65)

for eps in QNICCP_EPSILONS:
    print(f"\n[ε_q = {eps:.3f}]")

    acc, prec, rec, f1 = evaluate_epsilon(
        basic_model, test_loader, centroids_basic, eps, "Basic"
    )
    results["Basic"]["acc"].append(acc)
    results["Basic"]["prec"].append(prec)
    results["Basic"]["rec"].append(rec)
    results["Basic"]["f1"].append(f1)
    print(f"  Basic | Acc:{acc:.4f} Prec:{prec:.4f} Rec:{rec:.4f} F1:{f1:.4f}")

    acc, prec, rec, f1 = evaluate_epsilon(
        ades_model, test_loader, centroids_ades, eps, "ADES"
    )
    results["ADES"]["acc"].append(acc)
    results["ADES"]["prec"].append(prec)
    results["ADES"]["rec"].append(rec)
    results["ADES"]["f1"].append(f1)
    print(f"  ADES  | Acc:{acc:.4f} Prec:{prec:.4f} Rec:{rec:.4f} F1:{f1:.4f}")


# ── 11. SUMMARY TABLE ───────────────────────────────────────────

print(f"\n{'═'*80}")
print(f"  SUMMARY — QNI-CCP ε_q SWEEP  |  Malimg  |  Basic vs ADES")
print(f"{'═'*80}")
print(f"  {'ε_q':>6}   {'Basic Acc':>9} {'ADES Acc':>9}   "
      f"{'Basic F1':>9} {'ADES F1':>9}   {'Δ Acc':>8} {'Δ F1':>8}")
print(f"  {'─'*75}")

for i, eps in enumerate(QNICCP_EPSILONS):
    b_acc = results["Basic"]["acc"][i]
    a_acc = results["ADES"]["acc"][i]
    b_f1  = results["Basic"]["f1"][i]
    a_f1  = results["ADES"]["f1"][i]
    d_acc = a_acc - b_acc   # Positive = ADES more robust
    d_f1  = a_f1  - b_f1
    s_acc = "↑" if d_acc >= 0 else "↓"
    s_f1  = "↑" if d_f1  >= 0 else "↓"
    marker = " ◄ paper ε_q" if abs(eps - 0.027) < 1e-4 else ""
    print(f"  {eps:>6.3f}   {b_acc:>9.4f} {a_acc:>9.4f}   "
          f"{b_f1:>9.4f} {a_f1:>9.4f}   "
          f"{d_acc:>+7.4f}{s_acc} {d_f1:>+7.4f}{s_f1}{marker}")


# ── 12. PLOTS ───────────────────────────────────────────────────

metrics      = ["acc",      "prec",      "rec",    "f1"]
metric_names = ["Accuracy", "Precision", "Recall", "F1-Score (Macro)"]
colors       = {"Basic": "#E74C3C", "ADES": "#2980B9"}

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle(
    "QNI-CCP ε_q Sweep: Basic QNN vs ADES QNI-CCP — Malimg Dataset\n"
    "Latent-space perturbation strength vs classification performance",
    fontsize=13, fontweight='bold'
)
axes = axes.flatten()

for idx, (metric, mname) in enumerate(zip(metrics, metric_names)):
    ax         = axes[idx]
    basic_vals = results["Basic"][metric]
    ades_vals  = results["ADES"][metric]

    ax.plot(QNICCP_EPSILONS, basic_vals, marker='o',
            color=colors["Basic"], label='Basic QNN',
            linewidth=2, markersize=6)
    ax.plot(QNICCP_EPSILONS, ades_vals, marker='s',
            color=colors["ADES"], label='ADES QNI-CCP',
            linewidth=2, markersize=6, linestyle='--')

    # Paper reference line — Malimg ε_q ≈ 0.027
    ax.axvline(x=0.027, color='gray', linestyle=':',
               linewidth=1.5, label='Paper ε_q=0.027')

    # Robustness gap — green where ADES is more robust than Basic
    ax.fill_between(QNICCP_EPSILONS, basic_vals, ades_vals,
                    alpha=0.08, color='green', label='Robustness gap')

    ax.set_title(mname, fontsize=11, fontweight='bold')
    ax.set_xlabel("QNI-CCP Epsilon (ε_q)", fontsize=9)
    ax.set_ylabel(mname, fontsize=9)
    ax.set_ylim(-0.05, 1.05)
    ax.legend(fontsize=8, loc='lower left')
    ax.grid(True, alpha=0.3)
    ax.set_xticks(QNICCP_EPSILONS)
    ax.tick_params(axis='x', rotation=45, labelsize=7)

plt.tight_layout()
plt.savefig("ades_epsilon_sweep_malimg.png", dpi=150, bbox_inches='tight')
plt.show()
print("✅ Plot saved: ades_epsilon_sweep_malimg.png")


# ── 13. SAVE RESULTS ────────────────────────────────────────────

with open("ades_epsilon_sweep_results_malimg.txt", "w") as f:
    f.write("ADES QNI-CCP EPSILON SWEEP RESULTS — MALIMG DATASET\n")
    f.write("="*80 + "\n\n")
    f.write(f"{'eps_q':>6}  {'Model':>10}  "
            f"{'Accuracy':>10} {'Precision':>10} "
            f"{'Recall':>10} {'F1':>10}\n")
    f.write("-"*65 + "\n")
    for i, eps in enumerate(QNICCP_EPSILONS):
        for key, lbl in [("Basic", "Basic"), ("ADES", "ADES")]:
            f.write(f"{eps:>6.3f}  {lbl:>10}  "
                    f"{results[key]['acc'][i]:>10.4f} "
                    f"{results[key]['prec'][i]:>10.4f} "
                    f"{results[key]['rec'][i]:>10.4f} "
                    f"{results[key]['f1'][i]:>10.4f}\n")

print("✅ Results saved: ades_epsilon_sweep_results_malimg.txt")
```
